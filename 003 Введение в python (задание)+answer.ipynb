Нужно расшифровать химические символы в название химических элементов.
Для удобства, прилагается json файл, который ставит в соответствие химическому символу его химическое название.
Как прочитать этот файл в словарь python (dict):

periodic_table = json.load(open('periodic_table.json'))

Обратите внимание, что, например, "Bi" - это "Висмут", а не "БорЙод".
**То есть регистр (заглавные или прописные) букв имеет значение.**

#### Итоговое решение

```python
#vibo: загружаем библиотеки
import re
import json

#vibo: читаем данные из заданного файла
import_file_3 = open ('import_file_3.txt', 'r')
import_object_3 = import_file_3.read()

#vibo: разбиваем строку по буквам с верхним регистром
import_list_3 = re.findall(r'[A-Z]?[^A-Z]*', import_object_3)
#vibo: удаляем последний элемент списка (почему-то появлялся - '')
import_list_3 = import_list_3[:int(len(import_list_3))-1]

#vibo: импортирируем словарь
periodic_table = json.load(open('periodic_table.json'))

#vibo: делаем пустой список для накопления результата
result_list = []
#vibo: идем по элементам списка из файла
for element in import_list_3:
    #vibo: иммем один элемент, проходим по ключам словаря
    for key in periodic_table:
        #vibo: если ключ словаря совпадает с элементом списка
        if key == element:
            #vibo: добавляем элемент в итоговый лист
            result_list.append(periodic_table[key])

#vibo: преобразуем каждый элемент list в str и склеиваем join
result_str = ''.join([str(_) for _ in result_list])

#vibo: записываем ответ в файл
with open('v_output_file_3.txt', 'w') as f:
    f.write(result_str)
```

##### Код для ods.ai

```python
import re
import json

import_file_3 = open ('import_file_3.txt', 'r')
import_object_3 = import_file_3.read()

import_list_3 = re.findall(r'[A-Z]?[^A-Z]*', import_object_3)
import_list_3 = import_list_3[:int(len(import_list_3))-1]

periodic_table = json.load(open('periodic_table.json'))

result_list = []
for element in import_list_3:
    for key in periodic_table:
        if key == element:
            result_list.append(periodic_table[key])

result_str = ''.join([str(_) for _ in result_list])

with open('v_output_file_3.txt', 'w') as f:
    f.write(result_str)
```
